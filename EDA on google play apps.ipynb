```python
def clean_size_column(x):
    if 'M' in x:
        return float(x.replace('M',''))
    elif 'Varies with device' in x:
        return 'Varies with device'
    elif 'k' in x:
        return (float(x.replace('k',''))/1000)
    elif 'G' in x:
        return (float(x.replace('G',''))*1000)
        
df['Size'] = df['Size'].apply(clean_size_column)
```
